# Spadochroniarz: równania ruchu

## 1. Wstęp

Zajmiemy się problemem jakim jest opis ruchu spadochroniarza, a co za tym idzie jego położenia, prędkości i przyspieszenia. Począwszy od przypomnienia podstawowych wzorów fizycznych, przez ulepszenia modelu o coraz to nowe założenia, kończąc na finalnym modelu, pokażemy że pozornie proste zagadnienie posiada wiele czynników, które wpływają znacznie na rozwiązania. Zamysł niniejszego opracowania to ilustracja kroków jakie należy podjąć rozwiązując dowolny skomplikowany problem występujący w świecie fizycznym.

## 2. Spadek swobodny

Zaczniemy od wyprowadzenia modelu uwzględniającego jedynie siłę ciężkości punktu materialnego poruszającego się w próżni w pobliżu Ziemi, a co za tym idzie pominiemy opory powietrza. Wyprowadzimy w ten sposób najprostszy model, który będziemy w stanie poprawiać dodając kolejne założenia. Nie będziemy brać pod uwagę momentu rozłożenia spadochronu ze względu na brak oporu powietrza, ponieważ bez sił oporu nie wpłynie on na położenie. Korzystając więc z $II$ zasady dynamiki Newtona danej jako:

$$\LARGE{F = ma,}$$

Wiemy, że siła $F$ będzie w tym wypadku równa $F_g$, która z kolei jest równa $-mg$ (ze względu na przeciwny zwrot wektora przyspieszenia względem pionowej osi), gdzie $m$ to masa spadochroniarza wraz ze spadochronem, a $g$ to przyspieszenie ziemskie wynoszące $9,81\frac{m}{s^2}$ w pobliżu powierzchni Ziemi. Przyspieszenie spadochroniarza jest oznaczone poprzez $a$. Przyrównując do siebie te dwie równości i dzieląc przez masę, otrzymamy $a = -g$. Przyspieszenie jest natomiast drugą pochodną położenia po czasie. Zatem otrzymujemy równanie różniczkowe zwyczajne drugiego rzędu $\frac{d^2 x}{d t^2} = -g$. Całkując obustronnie dostaniemy:

$$v(t) = \frac{dx}{dt} = -gt + C,$$
Stała $C$ będzie pochodną położenia po czasie w chwili $t_0 = 0$, zatem jest to prędkość początkowa, którą oznaczymy przez $v_0$, co da:
$$v(t) = -gt + v_0,$$
Równanie to można ponownie scałkować otrzymując:
$$x(t) = -\frac{1}{2}\, gt^2 + v_0 t + C,$$
Podobnie jak w poprzednim równaniu stałą $C$ oznaczymy jako $x_0$, co będzie reprezentować położenie w chwili $t_0 = 0$. Finalnie wzór na położenie dla spadku swobodnego wynosi:
$$x(t) = -\frac{1}{2}\, gt^2 + v_0 t + x_0.$$

Wraz z warunkami początkowymi dla których $v_0 = 0, \ x_0=h$, mamy $x(t) = h -\frac{1}{2}\ gt^2$. Rozwiązanie to jednak nie brało pod uwagę oporów powietrza, które jak wiemy znacznie wpłyną na model. Spadochron spowalnia nas abyśmy mogli bezpiecznie wylądować, co nie jest brane pod uwagę w tym przypadku. Możemy jednak zauważyć, że masa spadochroniarza nie wpływa na jego położenie, dla takich założeń. Zatem w próżni spadochron nie wpływa w żaden sposób na równania ruchu.

## 3. Model uwzględniający opory powietrza

Kolejnym założeniem zwiększającym dokładność modelu jest uwzględnienie oporów podczas spadku. Problem ponownie sprowadzimy do zmiany położenia pionowego $x$, lub zmiany prędkości (względem osi pionowej) $v$, pod wpływem sił grawitacji i oporu powietrza. Zakładamy natomiast, że opór powietrza jest liniowo proporcjonalny do prędkości. Naturalne jest zauważenie, że wraz ze wzrostem prędkości wzrasta również siła oporu. Liniowość jest natomiast uzasadniona przez łatwość obliczeń. Do wyliczenia rzeczywistego wykładnika prędkości potrzebujemy jednak danych eksperymentalnych.

Korzystamy z $II$ zasady dynamiki Newtona, $F = ma$, gdzie spadochroniarz posiada masę $m$ oraz przyspieszenie $a$. Natomiast siła $F$ jest sumą siły grawitacji $F_g$ i siły oporu $F_d$. Jak wiemy z poprzedniego przypadku przyspieszenie jest pochodną prędkości po czasie, a prędkość to pochodna położenia po czasie, zatem $a = \frac{dv}{dt} = \frac{d^2 x}{d t^2}$. Siła grawitacji dana jest wzorem $F_g = -mg$, oraz siła oporu wyraża się poprzez $F_d = -kv$, gdzie $k$ to współczynnik proporcjonalności. Początkowo uwzględnimy dwa stany gdzie $k_1$ to współczynnik przed otwarciem spadochronu, a $k_2$ kiedy spadochron zostanie już otworzony. Jeśli zatem do otwarcia spadochronu dojdzie w chwili $t_0$, mamy dwa przypadki:

$$
k = \left\{
    \begin{aligned}
    &k_1, &0& \leq t < t_0 \\
    &k_2, &t& \geq t_0.
    \end{aligned}
\right.
$$

Rozwiążemy zatem równanie różniczkowe zwyczajne, pierwszego rzędu dla prędkości dane poprzez przyrówanania siły całkowitej oraz sum sił grawitacji i oporu:
$$m \frac{dv}{dt} = -mg -kv,$$

### 3.1 Spadek przed otworzeniem spadochronu

Dzieląc całe równanie przez $m$ wraz z $k = k_1$, otrzymamy równanie, które rozwiążemy używając czynnika całkującego $e^{\frac{k_1 t}{m}}$, co daje:

$$v' \ e^{\frac{k_1 t}{m}} = -g \ e^{\frac{k_1 t}{m}} - v \frac{k_1}{m} \ e^{\frac{k_1 t}{m}},$$
$$v' \ e^{\frac{k_1 \ t}{m}} + v \frac{k_1}{m} \ e^{\frac{k_1 t}{m}} = -g \ e^{\frac{k_1 t}{m}},$$
$$\int \left(v \ e^{\frac{k_1 t}{m}} \right)'\,dt = \int -g \ e^{\frac{k_1 t}{m}}\,dt,$$
$$v \ e^{\frac{k_1 t}{m}} = -\frac{mg}{k_1} \ e^{\frac{k_1 t}{m}} + C.$$

Z założeń początkowych, $v(0) = 0$, stała $C$ jest równa $\frac{mg}{k_1}$. Upraszczając równanie oraz korzystając z wyznaczonej stałej otrzymamy:
$$\begin{aligned}
v(t) = \frac{mg}{k_1} \left(e^{\frac{-k_1 t}{m}} - 1 \right), && 0 \leq t < t_0.
\end{aligned}
$$

Położenie otrzymamy poprzez nałożenie całki na obie strony tego równania: 
$$\begin{aligned}
x(t) &= \int v(t) \,dt, \\
x(t) &= \int -\frac{mg}{k_1} + \frac{mg}{k_1} \ e^{\frac{-k_1 t}{m}} \,dt, \\
x(t) &= -\frac{mg}{k_1} \ t - \frac{m^2 g}{k_1^2} \ e^{\frac{-k_1 t}{m}} + C.
\end{aligned}
$$

Z warunków początkowych, $x(0) = h$, otrzymujemy stałą $C = h + \frac{m^2 g}{k_1^2}$, zatem wzór na położenie to:
$$\begin{aligned}
x(t) = h -\frac{mg}{k_1} \ t - \frac{m^2 g}{k_1^2} \left(e^{\frac{-k_1 t}{m}} - 1 \right), && 0 \leq t < t_0.
\end{aligned}
$$

### 3.2 Rozłożenie spadochronu

Po chwili $t_0$ następuje rozłożenie spadochronu, co oznacza że zmienia się nasz współczynnik proporcjonalności w sile oporu. Zatem $k = k_2$. Rozwiązanie jednak nie będzie się znacznie różniło. Korzystając z:
$$\begin{aligned}
m \frac{dv}{dt} = -mg - k_2 v, && v(t) = \frac{mg}{k_1} \left(e^{\frac{-k_1 t}{m}} - 1 \right), \text{dla} && 0 \leq t < t_0.
\end{aligned}
$$

Ponownie poprzez podzielenie pierwszego równania przez $m$ i pomnożenie przez czynnik całkujący, który w tym wypadku wynosi $e^{\frac{k_2 t}{m}}$, a później nałożenie całki na obie strony równania, otrzymujemy:

$$v' \ e^{\frac{k_2 t}{m}} = -g \ e^{\frac{k_2 t}{m}} - v \frac{k_2}{m} \ e^{\frac{k_2 t}{m}},$$
$$v' \ e^{\frac{k_2 \ t}{m}} + v \frac{k_2}{m} \ e^{\frac{k_2 t}{m}} = -g \ e^{\frac{k_2 t}{m}},$$
$$\int \left(v \ e^{\frac{k_2 t}{m}}\right)'\,dt = \int -g \ e^{\frac{k_2 t}{m}}\,dt,$$
$$v \ e^{\frac{k_2 t}{m}} = -\frac{mg}{k_2} \ e^{\frac{k_2 t}{m}} + C,$$

Więc dla czasu $t \geq t_0$ prędkość jest dana wzorem:
$$v(t) = -\frac{mg}{k_2} + C \ e^{\frac{-k_2 t}{m}}.$$

Stałą wyznaczymy ze wzoru dla $t < t_0$, ponieważ funkcja opisująca prędkość musi być ciągła, a co za tym idzie $\lim\limits_{t \to t_0} v(t) = v(t_0)$, zatem:

$$\frac{mg}{k_1} \left(e^{\frac{-k_1 t_0}{m}} - 1 \right) = -\frac{mg}{k_2} + C \ e^{\frac{-k_2 t_0}{m}},$$
$$\frac{mg}{k_1} \left(e^{\frac{-k_1 t_0}{m}} - 1 \right) + \frac{mg}{k_2} = C \ e^{\frac{-k_2 t_0}{m}},$$
$$C = \frac{mg}{k_1} \left(e^{\frac{-k_1 t_0}{m}} - 1 \right) e^{\frac{k_2 t_0}{m}}  + \frac{mg}{k_2} \ e^{\frac{k_2 t_0}{m}}.$$

I podstawiając $C$ do wzoru na prędkość:
$$v = -\frac{mg}{k_2} + e^{\frac{-k_2 t}{m}} \left[\frac{mg}{k_1} \left(e^{\frac{-k_1 t_0}{m}}-1\right) e^{\frac{k_2 t_0}{m}} + \frac{mg}{k_2} \ e^{\frac{k_2 t_0}{m}}\right],$$
$$v = -\frac{mg}{k_2} + \frac{mg}{k_1} \left(e^{\frac{-k_1 t_0}{m}}-1\right) e^{\frac{k_2 (t_0-t)}{m}} + \frac{mg}{k_2} \ e^{\frac{k_2 (t_0-t)}{m}},$$
$$v = \frac{mg}{k_2}\left(e^{\frac{-k_2 (t-t_0)}{m}} - 1\right) + \frac{mg}{k_1} \left(e^{\frac{-k_1 t_0}{m}}-1\right) e^{\frac{-k_2 (t-t_0)}{m}},$$

Finalnie wzór na prędkość dla $t > t_0$ to:
$$v(t) = \frac{mg}{k_2}\left(e^{\frac{-k_2 (t-t_0)}{m}} - 1\right) + \frac{mg}{k_1} \left(e^{\frac{-k_1 t_0}{m}}-1\right) e^{\frac{-k_2 (t-t_0)}{m}}.$$

Całkjąc to równanie otrzymamy wzór na położenie po czasie $t_0$:
$$x(t) = \int \frac{mg}{k_2}\left(e^{\frac{-k_2 (t-t_0)}{m}} - 1\right) + \frac{mg}{k_1} \left(e^{\frac{-k_1 t_0}{m}}-1\right) e^{\frac{-k_2 (t-t_0)}{m}}\,dt,$$
$$x(t) = \frac{mg}{k_2} \int \left(e^{\frac{-k_2 (t-t_0)}{m}} - 1\right)\,dt + \frac{mg}{k_1}\left(e^{\frac{-k_1 t_0}{m}}-1\right) \int e^{\frac{-k_2 (t-t_0)}{m}}\,dt,$$
$$x(t) = -\frac{mg}{k_2} \ t -\frac{mg}{k_2} \left[\frac{m}{k_2} + \frac{m}{k_1}\left(e^{\frac{-k_1 t_0}{m}}-1\right) \right] e^{\frac{-k_2 (t-t_0)}{m}} + C,$$
$$x(t) = -\frac{mg}{k_2} \ t -\frac{m^2g}{(k_1 k_2)^2} \left[k_1^2 + k_1k_2\left(e^{\frac{-k_1 t_0}{m}}-1\right) \right] e^{\frac{-k_2 (t-t_0)}{m}} + C.$$

Ponownie stałą wyznaczymy z faktu, że położenie jest funkcją ciągłą, więc $\lim\limits_{t \to t_0} x(t) = x(t_0),$ co daje nam:
$$C = h - \frac{mg \ t_0 }{k_1 k_2} \ (k_2-k_1) + \frac{m^2g}{(k_1 k_2)^2} \left[k_1^2 + k_1k_2\left(e^{\frac{-k_1 t_0}{m}}-1\right) - k_2^2\left(e^{\frac{-k_1 t_0}{m}}-1\right) \right],$$
a w związku z tym:

$$x(t) = h - \frac{mg \ t_0 }{k_1 k_2} \ (k_2-k_1) -\frac{mg}{k_2}t + \frac{m^2g}{(k_1 k_2)^2} \left[k_1^2 + k_1k_2\left(e^{\frac{-k_1 t_0}{m}}-1\right) - k_2^2\left(e^{\frac{-k_1 t_0}{m}}-1\right) \right] -\frac{m^2g}{(k_1 k_2)^2} \left[k_1^2 + k_1k_2\left(e^{\frac{-k_1 t_0}{m}}-1\right) \right] e^{\frac{-k_2 (t-t_0)}{m}},$$
$$x(t) = h - \frac{mg}{k_1 k_2} (k_2t_0 + (t - t_0)) -\frac{m^2g}{(k_1 k_2)^2} \left[k_1^2 + k_1k_2\left(e^{\frac{-k_1 t_0}{m}}-1\right) \right] \left(e^{\frac{-k_2 (t-t_0)}{m}} - 1 \right) - \frac{m^2g}{k_1^2} \left(e^{\frac{-k_1 t_0}{m}}-1\right).$$

Łatwo można również obliczyć prędkość graniczną poprzez obserwację stanu obiektu który ją osiąga. Mianowicie, ze względu na definicje prędkości granicznej jest ona osiągana gdy $\frac{dv}{dt} = 0$. Wraz z zerową prędkością początkową otrzymamy $v_t = - \frac{mg}{k}$.

### 3.3 Wykresy przedstawiające wyniki

Zakładamy, że: wysokość skoku to $4000 m$, skoczek waży $70 kg$, jego prędkość początkowa wynosi $0 \frac{m}{s}$, siła oporu przed rozłożeniem spadochronu to $14 \frac{kg}{s}$, a po rozłożeniu $160 \frac{kg}{s}$, otwarcie spadochronu zostaje wykonane po $40 s$, przyspieszenie ziemskie wynosi $9.81 \frac{m}{s^2}$

```python
import math
import matplotlib.pyplot as plt
y0=4000
v0=0
m=70
op=14
op1=160
g=9.81
dt=0.01
t = 0
t0=40
y = y0
v = v0
positions = []
velocities = []
times = []
while True:
    y = y0 - (m*g)/op*t-(m*m*g)/(op*op)*(math.e**((-op*t)/m)-1) if t<t0 else (-m*g)/op1*t-(m*m*g)/(op*op1*op1)*(op+op1*(math.exp((-op*t0)/m)-1))*math.exp(-op1*(t-t0)/m)+y0-(m*g)/op*t0-(m*m*g)/(op*op)*(math.exp((-op*t0)/m)-1)+(m*g)/op1*t0+(m*g)/op1*(m/op1+m/op*(math.exp((-op*t0)/m)-1))
    v = (m*g)/op*(math.exp((-op*t)/m)-1) if t<t0 else (m*g)/op1*(math.exp(-op1*(t-t0)/m)-1)+(m*g)/op*(math.exp((-op*t0)/m)-1)*(math.exp((-op1*(t-t0)/m)))
    t += dt
    positions.append(y)
    velocities.append(v)
    times.append(t)
    if t>55:
        break
accelerations = [0] + [(velocities[i+1] - velocities[i])/dt for i in range(len(velocities)-1)]
fig, axs = plt.subplots(1,3, figsize=(18,4))
axs[0].plot(times, accelerations)
axs[0].set_title('przyspieszenie')
axs[0].set_xlabel('t [s]')
axs[0].set_ylabel('a [m/s^2]')
axs[1].plot(times, velocities)
axs[1].set_title('prędkość')
axs[1].set_xlabel('t [s]')
axs[1].set_ylabel('v [m/s]')
axs[2].plot(times, positions)
axs[2].set_title('droga')
axs[2].set_xlabel('t [s]')
axs[2].set_ylabel('x [m]')
plt.show()
```

Wykres pokazuje pierwsze $55$ sekund lotu spadochroniarza, w dalszym czasie wykresy znacznie się nie zmieniają ze względu na osiągnięcie prędkości granicznej.

## 4. Numeryczne rozwiązania modeli

W przedstawionych modelach założyliśmy, że spadochroniarz waży $70 kg$, wysokość z jakiej skaczemy to $4000 m$, spadochron zostaje otwarty po czasie $40 s$ od skoku, współczynniki oporu $k_0$ oraz $k_1$ mają wartości kolejno równe $14 \frac{kg}{s}$ oraz $160 \frac{kg}{s}$. Reszta założeń, takie jak stała grawitacyjna czy prędkość początkowa, zostają bez zmian. Stworzone wykresy pokazują zależności przyspieszenia, prędkości oraz położenia od czasu na przedziale czasowym $t \in (0,55 s)$.

### 4.1 Natychmiastowa zmiana współczynnika oporu dla $t_0$

Rozważymy model opisany w punkcie 3. Rozwiążemy go natomiast numerycznie aby później porównać metody z dokładnym wynikiem i wyznaczyć tę dokładniejszą.

#### Metoda Eulera

Schemat Eulera jest metodą numeryczną, która polega na przybliżeniu rozwiązania równania różniczkowego zwyczajnego poprzez założenie, że prędkość zmiany funkcji w danym punkcie jest równa przyrostowi funkcji w tym punkcie. Wzór schematu Eulera jest następujący:
$$
y_{n+1} = y_n + h f(x_n,y_n).
$$
Gdzie:
- $y_n$ jest wartością funkcji zależnej w punkcie $x_n$
- $h$ jest krokiem numerycznym
- $f(x_n,y_n)$ jest wartością pochodnej funkcji zależnej w punkcie $x_n$.

```python
def dvdt(t,v):
    if t<t0:
        return (-op*v - m*g)/m
    else:
        return ((-op1*v - m*g)/m)

def dydt(t,y,v):
    return v
```

```python
def euler(y0, v0):
  t=0
  y = y0
  v = v0
  positionse1 = []
  velocitiese1 = []
  times = []
  while y >0:
    a = dvdt(t,v)
    y += v*dt
    v += a*dt
    t += dt
    positionse1.append(y)
    velocitiese1.append(v)
    times.append(t)
    if t>55:
        break
  return positionse1, velocitiese1, times

positionse1, velocitiese1, times = euler(y0, v0)
accelerationse1 = [0] + [(velocitiese1[i+1] - velocitiese1[i])/dt for i in range(len(velocitiese1)-1)]

import matplotlib.pyplot as plt
fig, axs = plt.subplots(1,3, figsize=(18,4))
axs[0].plot(times, accelerationse1, color='orange')
axs[0].set_title('przyspieszenie')
axs[0].set_xlabel('t [s]')
axs[0].set_ylabel('a [m/s^2]')
axs[1].plot(times, velocitiese1, color='orange')
axs[1].set_title('prędkość')
axs[1].set_xlabel('t [s]')
axs[1].set_ylabel('v [m/s]')
axs[2].plot(times, positionse1, color='orange')
axs[2].set_title('droga')
axs[2].set_xlabel('t [s]')
axs[2].set_ylabel('x [m]')
plt.show()
```

#### Metoda Rungego-Kutty

Metoda Rungego-Kutty czwartego rzędu jest metodą numeryczną rozwiązywania równań różniczkowych zwyczajnych. Polega ona na dokładniejszym przybliżeniu pochodnej funkcji poprzez interpolację pochodnej funkcji w danym punkcie za pomocą pochodnych funkcji w kilku innych punktach. Wzór metody Rungego-Kutty czwartego rzędu jest następujący:
$$
\begin{aligned}
k_1 &= hf(x_n, \ y_n) \\
k_2 &= hf \left(x_n + \frac{h}{2}, \ y_n + \frac{k_1}{2} \right) \\
k_3 &= hf \left(x_n + \frac{h}{2}, \ y_n + \frac{k_2}{2} \right) \\
k_4 &= hf(x_n + h, \ y_n + k_3) \\
y_{n+1} &= y_n + \frac{1}{6} \, (k_1 + 2k_2 + 2k_3 + k_4).
\end{aligned}
$$
Gdzie:
- $y_n$ jest wartością funkcji zależnej w punkcie $x_n$
- $h$ jest krokiem numerycznym
- $f(x_n, \ y_n)$ jest wartością pochodnej funkcji zależnej w punkcie $x_n$

```python
def runge_kutta(y0, v0):
    t=0
    y = y0
    v = v0
    positionsr1 = []
    velocitiesr1 = []
    times = []
    while y >0:
        k1 = dvdt(t,v)
        l1 = dydt(t,y,v)
        k2 = dvdt(t+dt/2,v+k1*dt/2)
        l2 = dydt(t+dt/2,y+l1*dt/2,v+k1*dt/2)
        k3 = dvdt(t+dt/2,v+k2*dt/2)
        l3 = dydt(t+dt/2,y+l2*dt/2,v+k2*dt/2)
        k4 = dvdt(t+dt,v+k3*dt)
        l4 = dydt(t+dt,y+l3*dt,v+k3*dt)
        v += (k1+2*k2+2*k3+k4)*dt/6
        y += (l1+2*l2+2*l3+l4)*dt/6
        t += dt
        positionsr1.append(y)
        velocitiesr1.append(v)
        times.append(t)
        if t>55:
            break
    return positionsr1, velocitiesr1, times


positionsr1, velocitiesr1, times = runge_kutta(y0, v0)
accelerationsr1 = [0] + [(velocitiesr1[i+1] - velocitiesr1[i])/dt for i in range(len(velocitiesr1)-1)]

import matplotlib.pyplot as plt
fig, axs = plt.subplots(1,3, figsize=(18,4))
axs[0].plot(times, accelerationsr1, color='green')
axs[0].set_title('przyspieszenie')
axs[0].set_xlabel('t [s]')
axs[0].set_ylabel('a [m/s^2]')
axs[1].plot(times, velocitiesr1, color='green')
axs[1].set_title('prędkość')
axs[1].set_xlabel('t [s]')
axs[1].set_ylabel('v [m/s]')
axs[2].plot(times, positionsr1, color='green')
axs[2].set_title('droga')
axs[2].set_xlabel('t [s]')
axs[2].set_ylabel('x [m]')
plt.show()
```

### 4.2 Dokładniejszy opis sił oporu

Początkowo zakładaliśmy, że siła oporu jest proporcjonalna do prędkości. Dokładniejszym opisem tej siły jest proporcjonalność do kwadratu prędkości [1], co z $II$ zasady dynamiki Newtona, wraz z warunkami początkowymi daje:
$$ \begin{aligned}
m \frac{dv}{dt} = kv^2 - mg, && v(0)=0.
\end{aligned}
$$

W przedstawionych modelach założyliśmy, że: spadochroniarz waży $70 kg$, wysokość skoku z jakiego wykonywana jest czynność to $4000 m$, spadochron zostaje otwarty w czasie $40 s$ od skoku. Współczynniki oporu w tym przypadku $k_0$ oraz $k_1$ mają wartości kolejno równe $1 \frac{kg}{m}$ oraz $10 \frac{kg}{m}$. Stworzone wykresy pokazują zależności przyspieszenia, prędkości oraz położenia od czasu na przedziale czasowym $t \in (0,100s)$.

Przypisanie stałym odpowiednich wartości (do wszystkich metod numerycznych w podpunktach 4.2-4.3):

```python
y0=4000 #wysokość początkowa
v0=0 #prędkość początkowa
m=70 #masa skoczka
op=1 #współczynnik oporu k0
op1=10 #współczynnik oporu k1
g=9.81 #stała grawitacyjna
dt=0.001 #skok czasowy
t0=40 #moment otwarcia spadochronu
r=2 #czas otwierania spadochronu (patrz 4.3)
time=[] #lista przetrzymywanych czasów (patrz 4.3)
```

#### Metoda Rungego-Kutty czwartego rzędu.

Zdefiniowanie metody Rungego-Kutty:

```python
def runge_kutta(y0, v0):
  t=0
  y = y0
  v = v0
  positionsr1 = []
  velocitiesr1 = []
  times = []
  
  while True:
    k1 = (op*v**2 - m*g)/m if t<t0 else (op1*v**2 - m*g)/m
    k2 = (op*(v+dt*k1/2)**2 - m*g)/m if t<t0 else (op1*(v+dt*k1/2)**2 - m*g)/m
    k3 = (op*(v+dt*k2/2)**2 - m*g)/m if t<t0 else (op1*(v+dt*k2/2)**2 - m*g)/m
    k4 = (op*(v+dt*k3)**2 - m*g)/m if t<t0 else (op1*(v+dt*k3)**2 - m*g)/m
    v += dt*(k1+2*k2+2*k3+k4)/6
    y += v*dt
    t += dt
    positionsr1.append(y)
    velocitiesr1.append(v)
    times.append(t)
    if t>100:
        break
  return positionsr1, velocitiesr1, times

positionsr1, velocitiesr1, times = runge_kutta(y0, v0)
accelerationsr1 = [0] + [(velocitiesr1[i+1] - velocitiesr1[i])/dt for i in range(len(velocitiesr1)-1)]
```

Rysowanie wykresu do zadanej metody:

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1,3, figsize=(18,4))
axs[0].plot(times, accelerationsr1)
axs[0].set_title('przyspieszenie')
axs[0].set_xlabel('t [s]')
axs[0].set_ylabel('a [m/s^2]')
axs[1].plot(times, velocitiesr1)
axs[1].set_title('prędkość')
axs[1].set_xlabel('t [s]')
axs[1].set_ylabel('v [m/s]')
axs[2].plot(times, positionsr1)
axs[2].set_title('droga')
axs[2].set_xlabel('t [s]')
axs[2].set_ylabel('x [m]')
plt.show()
```

Wykresy nie zmieniły się znacznie poza ich skalą, co bardziej odzwierciedla rzeczywistość.

### 4.3 Liniowa zmiana współczynnika oporu na przediale $(t_0,t_0 + r)$ 

Motywacją tej zmiany jest uwzględnienie czasu jaki spadochron potrzebuje aby został w pełni rozłożony. Nie jest realistyczne bowiem aby prędkość zmieniała się punktowo z jednej wartości na inną, a co za tym idzie, aby przyspieszenie było nieskończone.

Tak jak w podtemacie 4.2 używamy równania:
$$ \begin{aligned}
m \frac{dv}{dt} = kv^2 - mg, && v(0)=0.
\end{aligned}
$$

Współczynnik oporu oznaczymy jako:
$$
k =
\left\{
    \begin{aligned}
    & k_0, &0& \leq t < t_0 \\
    & k_0 + \frac{k_1-k_0}{r} \, (t-t_0), &t_0& \leq t < t_0 + r \\
    & k_1, &t& \geq t_0 + r.
    \end{aligned}
\right.
$$

```python
def op2(times):
    time.append(times[-2])
    time.append(times[-1])
    if((op+(op1-op)/r*(time[1]-time[0]))<=op1):
      x=op+(op1-op)/r*(time[1]-time[0])
      del time[1]
      return x
    else:
      return op1
```

```python
time=[]
def runge_kutta2(y0, v0):
  t=0
  y = y0
  v = v0
  positionsr2 = []
  velocitiesr2 = []
  times = []
  while True:
    k1 = (op*v**2 - m*g)/m if t<t0 else (op2(times)*v**2 - m*g)/m
    k2 = (op*(v+dt*k1/2)**2 - m*g)/m if t<t0 else (op2(times)*(v+dt*k1/2)**2 - m*g)/m
    k3 = (op*(v+dt*k2/2)**2 - m*g)/m if t<t0 else (op2(times)*(v+dt*k2/2)**2 - m*g)/m
    k4 = (op*(v+dt*k3)**2 - m*g)/m if t<t0 else (op2(times)*(v+dt*k3)**2 - m*g)/m
    v += dt*(k1+2*k2+2*k3+k4)/6
    y += v*dt
    t += dt
    positionsr2.append(y)
    velocitiesr2.append(v)
    times.append(t)
    if t>100:
        break
  return positionsr2, velocitiesr2, times
positionsr2, velocitiesr2, times = runge_kutta2(y0, v0)
accelerationsr2 = [0] + [(velocitiesr2[i+1] - velocitiesr2[i])/dt for i in range(len(velocitiesr2)-1)]

fig, axs = plt.subplots(1,3, figsize=(18,4))
axs[0].plot(times, accelerationsr2)
axs[1].plot(times, velocitiesr2)
axs[2].plot(times, positionsr2)
axs[0].set_title('przyspieszenie')
axs[0].set_xlabel('t [s]')
axs[0].set_ylabel('a [m/s^2]')
axs[1].set_title('prędkość')
axs[1].set_xlabel('t [s]')
axs[1].set_ylabel('v [m/s]')
axs[2].set_title('droga')
axs[2].set_xlabel('t [s]')
axs[2].set_ylabel('x [m]')
plt.show()
```

Widzimy zatem brak "ostrego" zwiększenia przyspiesznia, a co za tym idzie wyznaczyliśmy dokładniejszy model.

## 5. Dokładniejszy opis ośrodka oraz obiektu spadającego

Zajmiemy się teraz uwzględnieniem fizyki płynów w naszym modelu. Równania Naviera-Stokesa opisują ruch ciała poruszającego się przez lepki płyn. Prędkość przemieszczania opisywana jest w nich wraz ze zmienną bezwymiarową $Re$ zwaną liczbą Reynoldsa. W ogólności jest ona równa $\frac{\rho d v}{\mu}$, gdzie $\rho$ to gęstość płynu, $d$ to wymiar charakterystyczny zagadnienia, $v$ to prędkość charakterystyczna płynu, a $\mu$ to lepkość dynamiczna płynu. Przez prędkość charakterystyczną płynu należy rozumieć prędkość uśrednioną odnoszącą się do całości przepływu, a nie wybraną prędkość lokalną. Wymiar charakterystyczny zagadnienia należy rozumieć jako wymiar zjawiska o charakterze pewnej odległości mającej bezpośredni wpływ na stateczność ruchu płynu. Gęstość płynu oraz jego lepkość dynamiczna są parametrami fizycznymi, istniejącymi niezależnie od jego ruchu i są one jednoznacznie określone dla danych substancji (poza pewnymi substancjami jak płyny nienewtonowskie). Kiedy $Re \ll 1$, siła związana z lepkością dominuje, a siła oporu jest proporcjonalna liniowo do prędkości. Dla $Re > 10^3$ dominują siły związane z ciśnieniem, co powoduje, że siła oporu jest proporcjonalna do kwadratu prędkości [4].

Aby określić, który model najlepiej opisuje naszą sytuacje związaną ze skoczkiem musimy wyliczyć liczbę Reynoldsa odpowiednią dla niego. Na typowych wysokościach, z których skacze się ze spadochronem, gęstość $\rho$ i lepkość dynamiczna powietrza $\mu$ są w przybliżeniu stałe [1], gdzie $\rho \approx 1 \ \frac{kg}{m^3}$, $\mu \approx 1,5 \times 10^{-5} \frac{kg}{m s}$. Jako prędkość charakterystyczną wykorzystamy $v_t$, czyli prędkość graniczną. Estymacja tej prędkości to $v \approx 5,3 \frac{m}{s}$ [3], dla typowych skoków z rozłożonym spadochronem, a $v_t \approx 45 \frac{m}{s}$ [3] dla spadku swobodnego. Pełnie rozłożony spadochron posiada przekrój równy $A \approx 44 m^2$, co daje $d \approx 7,5 m$ a, wpadochroniarz w pozycji spadku swobodnego $A \approx 0,5 m^2$, czyli $d \approx 0,8 m$. Można zatem wyliczyć, że liczba Reynoldsa $Re > 10^6$ dla sytuacji sprzed otworzenia spadochronu jak i po jego otworzeniu. Wynika z tego, że siła oporu jest prawidłowo modelowana, gdy jest proporcjonalna do kwadratu prędkości. Możemy zatem wyrazić siłę oporu jako [1]:
$$F_d = \frac{1}{2} \, (C_d A \rho v^2).$$

$C_d$ jako współczynnik oporu, jest wyliczany eksperymentalnie, wyczytamy go z tabeli podanej poniżej.

| Kształt | Liczba Reynoldsa | $C_d$ |
| --- | --- | --- |
| Półkulista powłoka | $Re > 10^3$ | $1,33$ |
| Dysk | $Re > 10^3$ | $1,10$ |
| Płaski pasek | $Re > 10^3$ | $1,95$ |
| Cylinder | $10^3 < Re < 2 \times 10^5$ | $1,95$ |
| Cylinder | $Re > 2 \times 10^5$ | $\approx 0,35$ |
| Sfera | $10^3 < Re < 2 \times 10^5$ | $0,45$ |
| Sfera | $Re > 3 \times 10^5$ | $\approx 0,20$ |

Siła oporu jest wytwarzana przez ciało skoczka, linie na których zawieszony jest spadochron oraz czaszę spadochronu. Istnieją różne metody rozłożenia spadochronu. Możliwe jest rozłożenie linii podtrzymujących czaszę jako pierwszych aby później w pełni rozłożyć spadochron. Sytuacja odwrotna jest również możliwa, jak i kombinacja rozłożenia linii oraz czaszy w tym samym momencie. My skupimy się na tej pierwszej metodzie. Posiada on trzy osobne fazy, jako pierwsze rozkładane są linie w czasie $t = t_0$, kiedy pociągnięta jest linka rozkładająca spadochron. W czasie $t = t_1$ spadochroniarz za pomocą siły oporu przechodzi z poziomej pozycji do pionowej, a czasza zaczyna się napełniać. Następną fazą jest pełne rozłożenie czaszy które trwa aż do czasu $t = t_2$. Pomiędzy czasem $t_2$, a $t_3$ pęd otaczającego powietrza nadmiernie rozciąga czaszę spadochronu, która powraca do swojej stabilnej, w pełni rozłożonej formy dla $t > t_3$. Podczas skoku zarówno ciało spadochroniarza jak i jego sprzęt generują siłę oporu, zatem całkowita siła oporu jest równa:
$$ F_d = F_d^b + F_d^e = \frac{1}{2} \, \rho \left(C_d^b A^b + C_d^e A^e \right)v^2.$$

Indeksy górne $b$ oraz $e$ oznaczają odpowiednio skoczka i jego sprzęt. Model ten ignoruje opory wygenerowane przez linie oraz zakłada, że są one nierozciągliwe. Nie jest to co prawda zgodne z rzeczywistością, lecz jest to marginalna różnica. Rozważać będziemy standardowy spadochron desantowy T-10, posiadający średnicę czaszy równą $d_0 = 10,7 m$. Kompletnie napełniony powietrzem zmniejsza swoją średnicę do $d_p = 7,47 m$, z przekrojem poziomym równym $a_1 = 43,8 m^2$. Liny utrzymujące spadochron mają długość około 84% średnicy $d_0$ tj. $l = 8,96 m$. Przeciętny skoczek skacząc z samolotu może być reprezentowany jako cylinder o wysokości $h = 1,78$, o przekroju poprzecznym $b_0 = 0,1 m^2$. Nie utrzymuje on jednak pionowej postawy podczas spadku swobodnego, ponieważ ta pozycja jest bardzo niestabilna i ciężko by było ją utrzymać. Przyjmuje on zatem pozycje poziomą, którą można zamodelować jako płaski pasek o przekroju poprzecznym równym $b_1 = 0,5 m^2$. Spadochron wraz z linami waży około $6,3 kg$, uprząż to kolejne $4,54 kg$, a spadochroniarz waży $70 kg$. Mamy zatem łączną masę równą $m = 6,3 + 4,54 + 70 = 80,54 kg$. Czas $t_0$ podczas którego zaczyna się rozkładanie spadochronu zaczyna się od chwili pociągnięcia linki albo na ustalonej wysokości. Zazwyczaj skacze się z wysokości około $x_0 \approx 1200 m$, a spadochron rozkładamy po $t_0 = 10 s$. Można by jednak otworzyć spadochron później lecz zostawiany jest margines błędu w celu otworzenia zapasowego spadochronu w przypadku awarii. Moment pełnego rozłożenia linii następuje w czasie $t_1 - t_0 = 0,5 s$ a rozłożenie spadochronu to czas $t_2 - t_1 = 1 s$. Całkowite rozłożenie zajmuje około $3,2 s$, zatem $t_3 - t_2 = 1,7 s$. Założymy też, że rozłożenie linii, a co za tym idzie zwiększenie ich długości następuje liniowo od czasu $t_0$ do $t_1$. Jednak dla zwiększania rozmiaru czaszy oznaczymy $A_{1,2}^e$ przez dane eksperymentalne przekroju poprzecznego dla czasu od $t_1$ do $t_2$. Następnie aż do $t_3$ następuje nadmierne napełnienie czaszy co podobnie zamodelujemy poprzez dane eksperymentalne oznaczone jako $A_{2,3}^e$. Zatem łącząc te wszystkie dane otrzymamy:

$$
\begin{aligned}
A^b(t) = \left\{
    \begin{aligned}
    &b_0, &0& \leq t < t_0 \\
    &b_0, &t_0& \leq t < t_1 \\
    &b_1, &t_1& \leq t < t_2 \\
    &b_1, &t_2& \leq t < t_3 \\
    &b_1, &t& \geq t_3
    \end{aligned}
\right.
&;&
C_d^b(t) = \left\{
    \begin{aligned}
    &1,95, &0& \leq t < t_0 \\
    &1,95, &t_0& \leq t < t_1 \\
    &0,35 \, h, &t_1& \leq t < t_2 \\
    &0,35 \, h, &t_2& \leq t < t_3 \\
    &0,35 \, h, &t& \geq t_3
    \end{aligned}
\right.
&;&
A^e(t) = \left\{
    \begin{aligned}
    &0, &0& \leq t < t_0 \\
    &b_1, &t_0& \leq t < t_1 \\
    &A_{1,2}^e(t), &t_1& \leq t < t_2 \\
    &A_{2,3}^e(t), &t_2& \leq t < t_3 \\
    &a_1, &t& \geq t_3
    \end{aligned}
\right.
&;&
C_d^e(t) = \left\{
    \begin{aligned}
    &0, &0& \leq t < t_0 \\
    &0,35\, \frac{t-t_0}{t_1-t_0}, &t_0& \leq t < t_1 \\
    &1,33, &t_1& \leq t < t_2 \\
    &1,33, &t_2& \leq t < t_3 \\
    &1,33, &t& \geq t_3.
    \end{aligned}
\right.
\end{aligned}
$$

## 6. Dokładniejszy model

Biorąc opis sił z modelu w podpunkcie 4.2:
$$ \begin{aligned}
m \frac{dv}{dt} = kv^2 - mg, && v(0)=0.
\end{aligned}
$$

Współczynnik oporu dany jest poprzez:
$$
k = \frac{1}{2} \, \rho \left(C_d^b A^b + C_d^e A^e \right) = \frac{1}{2} \, \rho \
\left\{
    \begin{aligned}
    & 1,95 \, b_0, &0& \leq t < t_0 \\
    & 1,95 \, b_0 + 0,35 \, b_1 l \ \frac{t-t_0}{t_1-t_0}, &t_0& \leq t < t_1 \\
    & 0,35 \, b_1 h + 1,33 \, A_{1,2}^e(t), &t_1& \leq t < t_2 \\
    & 0,35 \, b_1 h + 1,33 \, A_{2,3}^e(t), &t_2& \leq t < t_3 \\
    & 0,35 \, b_1 h + 1,33 \, a_1, &t& \geq t_3.
    \end{aligned}
\right.
$$

Aby rozwiązać to równanie analitycznie postępowalibyśmy podobnie jak dla modelu z drugiego podpunktu. Na początku należy rozwiązać równanie różniczkowe z warunkami początkowymi na przedziale od $0$ do $t_0$ i użyć tego rozwiązania aby wyliczyć stałą $C$ dla przedziału od $t_0$ do $t_1$, który wpierw wyliczymy analitycznie z dokładnością do stałej, ponieważ prędkość musi być funkcją ciągłą. Postępowalibyśmy podobnie dla kolejnych przedziałów. Obliczenia te są zbyt długie i mało informacyjne. Zatem użyjemy programu, który zaprezentuje rozwiązanie numeryczne, aby zwizualizować nasz model oraz aby dało się go porównać z poprzednim modelem. Możemy jednak wyliczyć warunki, dla których $k$ jako współczynnik oporu jest ciągły i różniczkowalny. Oczywistym jest, że na swoich przedziałach będzie on ciągły, ponieważ jest zadany funkcjami ciągłymi. Musimy zatem pokazać kiedy będzie on ciągły w punktach $t_0$, $t_1$, $t_2$ i $t_3$. Policzymy więc granice lewostronne w tych punktach i przyrównamy do odpowiadających im wartości funkcji:

$$
\begin{aligned}
\lim\limits_{t \to t_0} k(t) = k(t_0) &\Rightarrow& 1,95 \, b_0 &= 1,95 \, b_0, \\
\\
\lim\limits_{t \to t_1} k(t) = k(t_1) &\Rightarrow& 1,95 \, b_0 + 0,35 \, b_1 l &= 0,35 \, b_1 h + 1,33 \, A_{1,2}^e(t_1) \\
&& A_{1,2}^e(t_1) &= \frac{1,95 \, b_0 + 0,35 \, b_1 \ (l-h)}{1,33}, \\
\\
\lim\limits_{t \to t_2} k(t) = k(t_2) &\Rightarrow& 0,35 \, b_1 h + 1,33 \, A_{1,2}^e(t_2) &= 0,35 \, b_1 h + 1,33 \, A_{2,3}^e(t_2) \\
&& A_{1,2}^e(t_2) &= A_{2,3}^e(t_2), \\
\\
\lim\limits_{t \to t_3} k(t) = k(t_3) &\Rightarrow& 0,35 \, b_1 h + 1,33 \, A_{2,3}^e(t_3) &=  0,35 \, b_1 h + 1,33 \, a_1\\
&& A_{2,3}^e(t_3) &= a_1.
\end{aligned}
$$

Dane eksperymentalne [2] wskazują na to, że po otwarciu spadochronu przekrój poziomy spadochronu zwiększa się wykładniczo w następujący sposób:
$$A_{1,2}^e(t) = \alpha _0 \, e^{\beta _0 \frac{t-t_1}{t_2-t_1}}.$$

Z poprzednich warunków możemy wyliczyć:
$$\begin{aligned}
\alpha _0 = \frac{1,95 \, b_0 + 0,35 \, b_1 (l - h)}{1,33};
&& \beta _0 = \ln{ \frac{a_1}{\alpha _0}}.
\end{aligned}
$$

Natomiast wzór opisujący, przekrój poziomy po przepełnieniu czaszy, zaproponowany w [1], który spełnia nasze warunki to:
$$ A_{2,3}^e(t) = a_1 \left( 1 + \beta _1 \sin{\left( \pi \, \frac{t-t_2}{t_3-t_2} \right)} \right),$$
gdzie $\beta _1$ reprezentuje wzrost w rozmiarze czaszy. Typowo spadochron powieksza się o 15% [1]. Zatem $\beta _1 \approx 0,15$.

Możemy zatem wyliczyć teraz numerycznie podane zagadnienie. Podstawiamy odpowiednie wartości do naszych stałych:

```python
import math
y0=4000
v0=0
dt=0.01
t0=10
t1=10.5
t2=11.5
t3=13.2
l=8.96
h=1.78
m=80.54
g=9.81
ro=1
b0=0.5
b1=0.1
a1=43.8
beta1=0.15
alfa0=(1.95*b0+0.35*b1*(l-h))/1.33
beta0=math.log(a1/alfa0)
```

Następnie możemy zdefiniować nasz współczynnik k i pokazać na wykresie.

```python
def k(t):
    A12t=alfa0*math.exp(beta0*(t-t1)/(t2-t1))
    A23t=a1*(1+beta1*math.sin(math.pi*(t-t2)/(t3-t2)))
    if t<t0:
        return 1/2*ro*1.95*b0
    if t0<=t<t1:
        return 1/2*ro*(1.95*b0+0.35*b1*l*(t-t0)/(t1-t0))
    if t1<=t<t2:
        return 1/2*ro*(0.35*b1*h+1.33*A12t)
    if t2<=t<t3:
        return 1/2*ro*(0.35*b1*h+1.33*A23t)
    if t>=t3:
        return 1/2*ro*(0.35*b1*h+1.33*a1)

wspolczynnik=[]
czas=[]
for i in range(0,2000):
    wspolczynnik.append(k(i/100))
    czas.append(i/100)

import matplotlib.pyplot as plt
plt.plot(czas, wspolczynnik)
plt.title("wartość k w zależności od czasu")
plt.show()
```

Podczas porównywania metod obliczyliśmy, że dokładniejszą metodą jest metoda Rungego-Kutty, więc używamy go do naszego zagadnienia:

```python
def dvdt(t,v):
    return (k(t)*v**2 - m*g)/m
def dydt(t,y,v):
    return v

def runge_kutta(y0, v0):
    t=0
    y = y0
    v = v0
    positionsr = []
    velocitiesr = []
    times = []
    while y >0:
        k1 = dvdt(t,v)
        l1 = dydt(t,y,v)
        k2 = dvdt(t+dt/2,v+k1*dt/2)
        l2 = dydt(t+dt/2,y+l1*dt/2,v+k1*dt/2)
        k3 = dvdt(t+dt/2,v+k2*dt/2)
        l3 = dydt(t+dt/2,y+l2*dt/2,v+k2*dt/2)
        k4 = dvdt(t+dt,v+k3*dt)
        l4 = dydt(t+dt,y+l3*dt,v+k3*dt)
        v += (k1+2*k2+2*k3+k4)*dt/6
        y += (l1+2*l2+2*l3+l4)*dt/6
        t += dt
        positionsr.append(y)
        velocitiesr.append(v)
        times.append(t)
        if t>55:
            break
    return positionsr, velocitiesr, times

positionsr, velocitiesr, times = runge_kutta(y0, v0)
accelerationsr = [(velocitiesr[1] - velocitiesr[0])/dt] + [(velocitiesr[i+1] - velocitiesr[i])/dt for i in range(len(velocitiesr)-1)]
```

Wykres dla naszego zagadnienia będzie wyglądał następująco:

```python
fig, axs = plt.subplots(1,3, figsize=(18,4))
axs[0].plot(times, accelerationsr)
axs[0].set_title('przyspieszenie')
axs[0].set_xlabel('t [s]')
axs[0].set_ylabel('a [m/s^2]')
axs[1].plot(times, velocitiesr)
axs[1].set_title('prędkość')
axs[1].set_xlabel('t [s]')
axs[1].set_ylabel('v [m/s]')

axs[2].plot(times, positionsr)

axs[2].set_title('droga')
axs[2].set_xlabel('t [s]')
axs[2].set_ylabel('x [m]')
plt.show()
```
